# 2.0 Data Wrangling

```python
def cleaning_data(table):
    
    #cleaned_soup = BeautifulSoup( table, 'html.parser')
    cleaned_soup = str(table)
    cleaned_soup = cleaned_soup.split('\n')
    
    return cleaned_soup
```

```python
def data_wrangling():
    
    list_scrapped = [fifa_2018_page1, fifa_2018_page2, fifa_2014_page1, fifa_2014_page2, fifa_2010_page1, fifa_2010_page2,
                     fifa_2006_page1, fifa_2006_page2, fifa_2002_page1, fifa_2002_page2, fifa_1998_page1, fifa_1998_page2,
                     fifa_1994_page1, fifa_1994_page2]
    
    all_ratings = [cleaning_data(c) for c in list_scrapped]
    
    ratings_cleaned = []
    for i in range(len(all_ratings)):
        ratings_cleaned.append(all_ratings[i][6:])
    
    ratings_cleaned = sum(ratings_cleaned, [])
    
    rankings = ratings_cleaned[0::4]
    teams = ratings_cleaned[1::4]
    ratings = ratings_cleaned[2::4]
    
    columns = ['rank', 'team', 'rating']
    
    df = pd.DataFrame(list(zip(rankings, teams, ratings)), columns=columns)
    df['year'] = 0
    
    df['year'][:100] = 2018
    df['year'][100:200] = 2014
    df['year'][200:300] = 2010
    df['year'][300:400] = 2006
    df['year'][400:500] = 2002
    df['year'][500:600] = 1998
    df['year'][600:] = 1994
    
    df['team'] = df['team'].str.lower()
    
    return df
```

```python
df_final = data_wrangling()
```
